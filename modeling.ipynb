```python
i = list_plant[0]

df_plant = df[df["Plant Id"] == i]
list_workorder = df_plant["Internal Work Order Id"].unique()
df_prefinish_drum_order = df_prefinish_drum[df_prefinish_drum["Site"] == i]
list_select_drum = list(df_prefinish_drum[df_prefinish_drum["Site"] == i].index)
RMI_inventory_order = RMI_inventory[RMI_inventory["Location Name"] == i]
list_select_RMI = list(RMI_inventory[RMI_inventory["Location Name"] == i].index)
order = 0
```

```python
for j in list_workorder:
    #each workorder
    df_order = df_plant[df_plant["Internal Work Order Id"] == j]
    list_color = df_order["Color"].unique()        
    
    for k in list_color:
        #filter color
        df_color = df_order[df_order["Color"] == k]
        df_split = df_classifier_split[df_classifier_split["Color"] == k]
        df_prefinish_drum_order_select = df_prefinish_drum_order[df_prefinish_drum_order["Color"] == "None"]
        list_select_drum_order = df_prefinish_drum_order_select.index[0:5]
        RMI_inventory_order_select = RMI_inventory_order[RMI_inventory_order["Color"] == k]
        list_select_RMI_order = RMI_inventory_order_select.index[0:1]

        #calculate the amount of RMI to classify
        df_pound = df_color[["Size","Flavor","Packaging Type","Qty"]]
        df_pound["Pound"] = (df_pound["Packaging Type"] == "Box")*df_pound["Qty"]*pound_per_box + (df_pound["Packaging Type"] == "Bag")*df_pound["Qty"]*pound_per_bag

        df_pound = df_pound[["Size","Pound"]].T

        df_pound.columns = df_pound.iloc[0,:]
        df_pound.drop("Size",axis = 0,inplace = True)
        df_pound = df_pound[["S1","S2","S3","S4","S5"]]

        df_split_1 = df_split[["Size","Percentage"]].T
        df_split_1.columns = df_split_1.iloc[0,:]
        df_split_1.drop("Size",axis = 0,inplace = True)
        df_split_1 = df_split_1[["S1","S2","S3","S4","S5"]]

        df_output = (df_pound.values/df_split_1.values).max()*df_split_1
        df_input = df_output.sum(axis=1)

        df_prefinish_drum_order_select.iloc[0:5,3] = df_output.values[0,:]
        df_prefinish_drum_order.loc[list_select_drum_order,"Capacity stored in pounds"] = df_prefinish_drum_order_select.iloc[0:5,3].values
        df_prefinish_drum_order.loc[list_select_drum_order,"Color"] = k
        df_prefinish_drum_order.loc[list_select_drum_order,"Size"] = np.array(["S1","S2","S3","S4","S5"])
        df_prefinish_drum_order.loc[list_select_drum_order,"Order"] = np.linspace(order,order+5-1,5)
        order += 5

        #reduce the RMI inventory
        RMI_inventory_order.loc[list_select_RMI_order,"Qty in pounds"] -= df_input.values 

#update prefinish drum
df_prefinish_drum_order.loc[list_select_drum,:] = df_prefinish_drum_order
```
